```python
# from profanityfilter import ProfanityFilter
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import numpy as np
import spacy

sia = SentimentIntensityAnalyzer()
# pf = ProfanityFilter()
nlp = spacy.load("en_core_web_sm")
```

```python
def text_selection(text):
  arr = text.split()
  s = set(arr)
  ln = len(arr)
  unics = len(s) / ln
  avg_word_len = np.mean([len(w) for w in s])

  # оч долго?
  # mat = sum(1 for word in arr if pf.is_profane(word))
  # mat_ratio = mat / ln

  chorus = arr.count('[Chorus]')

  sentiment = sia.polarity_scores(text)['compound']
  return {
      'num_words': ln,
      'unic_words_ratio': unics,
      'avg_word_len': avg_word_len,
      'chorus': chorus,
      'sentiment': sentiment
  }

```

```python
def desc_selection(desc):
  arr = text.split()
  ln = len(arr)
  avg_word_len = np.mean([len(w) for w in arr])
  doc = nlp(description)
  names = len(doc.ents)
  links = arr.count('http')
  return {
      'num_words': ln,
      'avg_word_len': avg_word_len,
      'names': names,
      'links': links
  }

```
